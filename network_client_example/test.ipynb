```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(100) / 100 *np.pi
y = x.copy()

plt.plot(x, y)
```

```python
y2 = 10 / (y + 0.2)

thresh = 15

plt.plot(x / np.pi * 180, y2, np.array([thresh]*len(x)))

print(y2.min(), y2.max())

mask = y2 > 5

print( sum(mask)) 

print(x[mask] / np.pi * 180)
print(y2[mask])
```
